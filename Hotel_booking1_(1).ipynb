### Variables Description

* **hotel                :** name of the hotel(City or Resort)
* **is_canceled      :** Whether the booking is canceled or not (0 for not canceled and 1 for canceled).
* **lead_time            :** number of days between the booking date and the arrival date.
* **arrival_date_year            :** Year of arrival date.
* **arrival_date_month           :** month of arrival date(Jan,Feb, etc).
* **arrival_date_week_number            :** week number of the arrival date(1 to 53).
* **arrival_date_day_of_month            :** day of the month of the arrival date(1 to 31).
* **stays_in_weekend_nights            :** Number of weekend nights(sat/sun) guest stayed or booked to stay at the hotel.
* **stays_in_week_nights            :** Number of week nights(mon to fri) the guest stayed or booked to stay ar the hotel.
* **adults            :** number of adults in booking record
* **children            :** number of children in booking record
* **babies            :** number of babies in booking record
* **meal            :** type of meal they booked(BB,HB,FB,etc)
* **country            :** country of origin of the guest.
* **market_segment            :** via what segment designation booking was made(direct,corporate etc).
* **distribution_channel            :** via which distribution_channel booking was made.
* **is_repeated_guest            :**  whether the booking was  made by a repeated guest(0 for No and 1 for Yes).
* **previous_cancellations            :** number of previous booking canceled by the guest.
* **previous_bookings_not_canceled            :**  number of previous bookings that were not canceled by the guest.
* **reserved_room_type            :** room type reserved by the guest.
* **assigned_room_type            :** room type assigned to the guest.
* **booking_changes            :** number of changes made to the booking by the guest.
* **deposit_type            :** type of deposit made for the booking (No deposit/ Refundable/ No refund)
* **agent            :** Id of agent for booking
* **days_in_waiting_list            :** number of days the booking was on the waiting list
* **customer_type            :** type of customer(Group,transient, transient-party,contract).
* **adr            :** the average daily rate, which is the average revenue per day for the booking.
* **required_car_parking_spaces            :** number of car parking asked in booking
* **total_of_special_requests            :** total number of special request made by the guest.
* **reservation_status            :** whether a customer has checked out or canceled,or not showed
* **reservation_status_date            :** date of making reservation status
